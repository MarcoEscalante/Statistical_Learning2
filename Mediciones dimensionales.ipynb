# Mediciones dimensionales

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

## Generamos Matrices

Desarrollamos una matriz de 5x2

```python
X = np.random.randn(5,2)
X
```

```python
X.shape
```

## Matriz con grado 2

Desarrollamos la matriz que tenemos en la primera parte y desarrollamos una nueva matriz con mayor tamaño, en este caso de grado 2.

```python
poly = PolynomialFeatures(degree=2)
X_grado2 = poly.fit_transform(X)
X_grado2
```

```python
X_grado2.shape
```

## Matriz con grado 3

Desarrollamos la matriz que tenemos en la primera parte y desarrollamos una nueva matriz con mayor tamaño, en este caso de grado 3.

```python
poly = PolynomialFeatures(degree=3)
X_grado3 = poly.fit_transform(X)
X_grado3
```

```python
X_grado3.shape
```

## Matriz con grado 10

Desarrollamos la matriz que tenemos en la primera parte y desarrollamos una nueva matriz con mayor tamaño, en este caso de grado 10.

```python
poly = PolynomialFeatures(degree=10)
X_grado10 = poly.fit_transform(X)
X_grado10
```

```python
X_grado10.shape
```
